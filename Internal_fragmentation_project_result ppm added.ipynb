```python
!pip install pyopenms

from pyopenms import AASequence, TheoreticalSpectrumGenerator, Param, MSSpectrum, Precursor

def calculate_isotopic_distribution(sequence):
    peptide = AASequence.fromString(sequence)

    # Create a TheoreticalSpectrumGenerator
    tsg = TheoreticalSpectrumGenerator()

    # Set precursor charge and isotope using Param
    param = Param()
    param.setValue("add_isotopes", "true")
    param.setValue("max_isotope", 2)
    param.setValue("min_isotope", 0)
    param.setValue("fragment_mass_tolerance", 0.01)
    tsg.setParameters(param)

    # Create an MSSpectrum for the precursor
    precursor = MSSpectrum()
    precursor.setMSLevel(1)
    precursor.setPrecursors([Precursor(peptide.getMonoWeight(), 1, 0)])  # Assuming charge 1

    # Generate the theoretical spectrum
    spec = MSSpectrum()
    tsg.getSpectrum(spec, peptide, precursor)

    return spec

def main(peptide_sequence):
    isotopic_distribution = calculate_isotopic_distribution(peptide_sequence)

    print(f"Isotopic Distribution for {peptide_sequence}:")
    print("Mass\tIntensity")
    for peak in isotopic_distribution:
        print(f"{peak.getMZ():.4f}\t{peak.getIntensity()}")

# Replace "YOUR_PEPTIDE_SEQUENCE" with the actual peptide sequence you want to analyze
peptide_sequence = "MGLSDGEWQQVLNVWGKVEADIAGHGQEVLIRLFTGHPETLEKFDKFKHLKTEAEMKASEDLKKHGTVVLTALGGILKKKGHHEAELKPLAQSHATKHKIPIKYLEFISDAIIHVLHSKHPGDFGADAQGAMTKALELFRNDIAAKYKELGFQG"
main(peptide_sequence)
```

```python
# Define amino acid masses with high resolution
aa_masses = {
    'A': 71.037114,
    'R': 156.101111,
    'N': 114.042927,
    'D': 115.026943,
    'C': 103.009185,
    'E': 129.042593,
    'Q': 128.058578,
    'G': 57.021464,
    'H': 137.058912,
    'I': 113.084064,
    'L': 113.084064,
    'K': 128.094963,
    'M': 131.040485,
    'F': 147.068414,
    'P': 97.052764,
    'S': 87.032028,
    'T': 101.047679,
    'W': 186.079313,
    'Y': 163.063329,
    'V': 99.068414
}

# Masses of water and ammonia molecules
H2O = 18.010564683
NH3 = 17.0265491015

def calculate_exact_mass(sequence, shift):
    """Calculates the exact mass of a protein sequence using high-resolution amino acid masses."""
    mass = sum(aa_masses[aa] for aa in sequence) + H2O + shift
    return mass

def get_internal_fragment_masses(sequence, ion_types):
    """Calculates internal fragment masses for the given ion types."""
    fragments = {}

    for ion_type in ion_types:
        shift = -H2O if ion_type in ['b', 'y'] else (-H2O - NH3 if ion_type == 'c' else -H2O + NH3 if ion_type == 'z' else 0)
        masses = []

        for i in range(len(sequence)):
            for j in range(i + 4, len(sequence)):
                subs = sequence[i:j+1]
                mass = calculate_exact_mass(subs, shift)
                masses.append(mass)

        fragments[ion_type] = masses

    return fragments

def main():
    sequence = input("Enter the protein sequence: ")
    fragmentation_type = input("Enter the fragmentation type (EThcD, ETD, CID, HCD): ")
    output_filename = input("Enter the name of the output TSV file (e.g., output.tsv): ")

    # Define ion types based on fragmentation method
    ion_types_map = {
        'EThcD': ['b', 'y', 'c', 'z'],
        'ETD': ['c', 'z'],
        'CID': ['b', 'y'],
        'HCD': ['b', 'y']
    }

    if fragmentation_type not in ion_types_map:
        print("Invalid fragmentation type. Please choose from EThcD, ETD, CID, HCD.")
        return

    ion_types = ion_types_map[fragmentation_type]
    fragment_data = get_internal_fragment_masses(sequence, ion_types)

    with open(output_filename, 'w') as file:
        file.write("Ion Type\tMass\n")
        for ion_type in fragment_data:
            masses = fragment_data[ion_type]
            for mass in masses:
                file.write(f"{ion_type.upper()}\t{mass:.4f}\n")

    print(f"Output written to {output_filename}")

if __name__ == "__main__":
    main()
```

```python
import pandas as pd

def match_masses(exp_file, insilico_file, tolerance=0.1):
    # Load CSV files
    exp_df = pd.read_csv(exp_file)
    insilico_df = pd.read_csv(insilico_file)

    # Ensure MonoisotopicMass column exists in both dataframes
    if 'MonoisotopicMass' not in exp_df.columns or 'MonoisotopicMass' not in insilico_df.columns:
        raise ValueError("MonoisotopicMass column not found in one or both files.")

    # Initialize dataframes for matched and unmatched masses
    matched_df = pd.DataFrame(columns=['MonoisotopicMass', 'Source'])
    exp_unmatched_df = pd.DataFrame(columns=['MonoisotopicMass'])
    insilico_unmatched_df = pd.DataFrame(columns=['MonoisotopicMass'])

    # Matching process
    for exp_mass in exp_df['MonoisotopicMass']:
        match_found = False
        for insilico_mass in insilico_df['MonoisotopicMass']:
            if abs(exp_mass - insilico_mass) <= tolerance:
                matched_df = matched_df.append({'MonoisotopicMass': exp_mass, 'Source': 'Both'}, ignore_index=True)
                match_found = True
                break
        if not match_found:
            exp_unmatched_df = exp_unmatched_df.append({'MonoisotopicMass': exp_mass}, ignore_index=True)

    for insilico_mass in insilico_df['MonoisotopicMass']:
        if not matched_df['MonoisotopicMass'].isin([insilico_mass]).any():
            insilico_unmatched_df = insilico_unmatched_df.append({'MonoisotopicMass': insilico_mass}, ignore_index=True)

    # Save results to files
    matched_df.to_csv('file_matched.csv', index=False)
    exp_unmatched_df.to_csv('exp_unmatched_file.csv', index=False)
    insilico_unmatched_df.to_csv('insilico_unmatched_file.csv', index=False)

    return "Matching completed. Files saved: file_matched.csv, exp_unmatched_file.csv, insilico_unmatched_file.csv"

# Example usage
match_masses('/content/OFCCF230704_22_ms2.csv', '/content/myoglobin 03_profile_ms2.csv', tolerance=0.1)
```

```python
##To find length of a file.
import pandas as pd

def count_rows_in_csv(file_paths):
    row_counts = {}
    for file_path in file_paths:
        try:
            # Open the file using pandas and calculate its length
            df = pd.read_csv(file_path)
            row_counts[file_path] = len(df)
        except Exception as e:
            # Handle exceptions (like file not found, invalid format, etc.)
            row_counts[file_path] = f"Error: {e}"

    return row_counts

# Example usage
file_paths = ['/content/exp_unmatched_file.csv']
result = count_rows_in_csv(file_paths)
print(result)
```

```python
#code for matching masses with theoratical internal ion generator
import pandas as pd

def compare_with_experiment(internal_ions_file, experimental_unmatched_masses, output_matched, output_unmatched, tolerance=0.1):
    # Load internal ions from CSV
    df_internal = pd.read_csv(internal_ions_file)

    # Print the column names in the internal ions file
    print("Internal Ions Column Names:", df_internal.columns)

    # Load experimental unmatched masses from CSV
    df_exp = pd.read_csv(experimental_unmatched_masses)

    # Print the column names in the experimental unmatched masses file
    print("Experimental Unmatched Masses Column Names:", df_exp.columns)

    # Match internal ions with experimental masses within the specified tolerance
    matches = df_exp[df_exp['MonoisotopicMass'].apply(lambda x: any(abs(x - ion) <= tolerance for ion in df_internal['Mass']))]

    # Unmatched masses
    unmatched = df_exp[~df_exp['MonoisotopicMass'].isin(matches['MonoisotopicMass'])]

    # Save matched and unmatched masses to CSV files
    matches.to_csv(output_matched, index=False)
    unmatched.to_csv(output_unmatched, index=False)

    print(f"Matched masses saved to {output_matched}")
    print(f"Unmatched masses saved to {output_unmatched}")

def main():
    # Specify the path to your internal ions CSV file
    internal_ions_file = "/content/Th_internal_ethcd.csv"

    # Specify the path to your experimental unmatched masses file
    experimental_unmatched_masses = "/content/exp_unmatched_file.csv"

    # Specify the output file paths
    output_matched = "matched_internal_ions.csv"
    output_unmatched = "unmatched_internal_ions.csv"

    # Compare internal ions with experimental unmatched masses
    compare_with_experiment(internal_ions_file, experimental_unmatched_masses, output_matched, output_unmatched)

if __name__ == "__main__":
    main()
```

```python
#####################################################
#script for matching masses based on ppm
import pandas as pd

def match_masses_and_count(exp_file, insilico_file, tolerance_ppm=5):
    # Load CSV files
    exp_df = pd.read_csv(exp_file)
    insilico_df = pd.read_csv(insilico_file)

    # Ensure MonoisotopicMass column exists in both dataframes
    if 'MonoisotopicMass' not in exp_df.columns or 'MonoisotopicMass' not in insilico_df.columns:
        raise ValueError("MonoisotopicMass column not found in one or both files.")

    # Initialize dataframes for matched and unmatched masses
    matched_df = pd.DataFrame(columns=['MonoisotopicMass', 'Source'])
    exp_unmatched_df = pd.DataFrame(columns=['MonoisotopicMass'])
    insilico_unmatched_df = pd.DataFrame(columns=['MonoisotopicMass'])

    # Function to calculate the tolerance in mass units based on ppm
    def calc_tolerance(mass):
        return mass * tolerance_ppm / 1e6

    # Matching process
    for exp_mass in exp_df['MonoisotopicMass']:
        match_found = False
        for insilico_mass in insilico_df['MonoisotopicMass']:
            if abs(exp_mass - insilico_mass) <= calc_tolerance(exp_mass):
                matched_df = matched_df.append({'MonoisotopicMass': exp_mass, 'Source': 'Both'}, ignore_index=True)
                match_found = True
                break
        if not match_found:
            exp_unmatched_df = exp_unmatched_df.append({'MonoisotopicMass': exp_mass}, ignore_index=True)

    for insilico_mass in insilico_df['MonoisotopicMass']:
        if not matched_df['MonoisotopicMass'].isin([insilico_mass]).any():
            insilico_unmatched_df = insilico_unmatched_df.append({'MonoisotopicMass': insilico_mass}, ignore_index=True)

    # Save results to files
    matched_df.to_csv('file_matched.csv', index=False)
    exp_unmatched_df.to_csv('exp_unmatched_file.csv', index=False)
    insilico_unmatched_df.to_csv('insilico_unmatched_file.csv', index=False)

    # Count the number of masses in each file
    count_matched = len(matched_df)
    count_exp_unmatched = len(exp_unmatched_df)
    count_insilico_unmatched = len(insilico_unmatched_df)

    return {
        "matched_count": count_matched,
        "exp_unmatched_count": count_exp_unmatched,
        "insilico_unmatched_count": count_insilico_unmatched
    }

# Example usagege
match_masses_and_count('/content/OFCCF231210_44_ms2.csv', '/content/Myo HCD 44_profile_ms2.csv', tolerance_ppm=5)
#print(result)
```
